```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
tb = pd.read_csv('StudentsPerformance.csv')
tb.head()
```

```python
tb.info()
```

```python
tb.duplicated().sum()
```

```python
tb["average_score"] = (tb["math score"] + tb["reading score"] + tb["writing score"] ) / 3
tb.head()
```

```python
tb.info()
```

```python
tb = tb.drop(columns=["math score", "reading score", "writing score"])
tb.head()
```

```python
y = tb["average_score"]
x = tb.drop(columns=["average_score"], axis=1)
x = pd.get_dummies(data=x, columns=["gender", "race/ethnicity", "parental level of education", "lunch", "test preparation course"], drop_first=True, dtype=int)
```

```python
x = sm.add_constant(x)
model = sm.OLS(endog=y, exog=sm.add_constant(x)).fit()
model.summary()
 
```
